# 01 Data Profiling
Load and explore the Amazon Reviews dataset structure, data types, and quality metrics.

```python
import pandas as pd
import numpy as np
import json
from pathlib import Path

# Configure paths
DATA_PATH = '../data/Amazon_Reviews.csv'
REPORT_PATH = '../assets/reports/data_profile.json'

print('Loading dataset...')
df = pd.read_csv(DATA_PATH, engine="python")
print(f'Shape: {df.shape}')
df.head()
```

```python
# Data types and missing values
print('Data Types:')
print(df.dtypes)
print('\nMissing Values:')
missing = df.isnull().sum()
print(missing[missing > 0] if missing.sum() > 0 else 'None')
print('\nDataFrame Info:')
df.info()
```

```python
# Basic statistics on numeric columns
print('Numeric Column Statistics:')
df.describe()
```

```python
# Column-specific profiling
if 'Rating' in df.columns:
    print('Rating Distribution:')
    print(df['Rating'].value_counts().sort_index())

if 'Review Text' in df.columns:
    print('\nText Length Statistics:')
    text_lens = df['Review Text'].fillna('').str.len()
    print(f'Mean: {text_lens.mean():.0f}, Min: {text_lens.min()}, Max: {text_lens.max()}')

if 'Review Title' in df.columns:
    print('\nReview Title Length Statistics:')
    title_lens = df['Review Title'].fillna('').str.len()
    print(f'Mean: {title_lens.mean():.0f}, Min: {title_lens.min()}, Max: {title_lens.max()}')
```

```python
# Generate data profile report
profile = {
    'shape': list(df.shape),
    'columns': list(df.columns),
    'dtypes': df.dtypes.astype(str).to_dict(),
    'missing_count': df.isnull().sum().to_dict(),
    'numeric_stats': df.describe().to_dict()
}

Path(REPORT_PATH).parent.mkdir(parents=True, exist_ok=True)
with open(REPORT_PATH, 'w') as f:
    json.dump(profile, f, indent=2, default=str)
    
print(f'Profile saved to {REPORT_PATH}')
```
